```python
# Create a Node class to create a node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Create a LinkedList class
class LinkedList:
    def __init__(self):
        self.head = None

    # Method to add a node at the beginning of the LL
    def insertAtBegin(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Method to add a node at the end of the LL
    def insertAtEnd(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next != None:
                current_node = current_node.next
            current_node.next = new_node

    # Method to add a node at a specific position
    def insertAtPosition(self, data, position):
        new_node = Node(data)
        if position == 1:
            new_node.next = self.head
            self.head = new_node
            return

        current_node = self.head
        current_position = 1
        while current_position < position - 1 and current_node != None:
            current_node = current_node.next
            current_position += 1
        if current_node == None:
            print("Invalid position")
            return
        new_node.next = current_node.next
        current_node.next = new_node

    # Method to delete a node from the beginning of a LL
    def deleteAtBegin(self):
        if self.head == None:
            print("Linked list is empty")
            return
        self.head = self.head.next

    # Method to delete a node from the end of the LL
    def deleteAtEnd(self):
        if self.head == None:
            print("Linked list is empty")
            return
        if self.head.next == None:
            self.head = None
            return

        current_node = self.head
        while current_node.next.next != None:
            current_node = current_node.next
        current_node.next = None

    # Mehod to delete a node from a specific position
    def deleteAtPosition(self, position):
        if self.head == None:
            print("Linked list is empty")
            return
        if position == 1:
            self.head = self.head.next
            return

        current_node = self.head
        current_position = 1
        while current_position < position - 1 and current_node != None:
            current_node = current_node.next
            current_position += 1

        if current_node == None or current_node.next == None:
            print("Invalid position")
            return
        current_node.next = current_node.next.next

    # Print the linked list
    def printLL(self):
        current_node = self.head
        while current_node != None:
            print(f"{current_node.data} -> ", end=" ")
            current_node = current_node.next
        print("None")

    # Methode to count size of LL
    def size(self):
        current_node = self.head
        count = 0
        while current_node != None:
            count += 1
            current_node = current_node.next
        return count

    # Method to update node at a given position
    def updateNode(self, val, index):
        current_node = self.head
        position = 1
        while current_node is not None and position != index:
            position += 1
            current_node = current_node.next

        if current_node is not None:
            current_node.data = val
        else:
            print("Index not present")

    # Method to reverse LL
    def reverseList(self):
      prev = None
      current = self.head
      while current:
          next_node = current.next
          current.next = prev
          prev = current
          current = next_node
      self.head = prev


# ===========================
# Menu-driven main program
# ===========================

if __name__ == '__main__':
    ll = LinkedList()

    while True:
        print("\n====== Linked List Menu ======")
        print("1. Insert at Beginning")
        print("2. Insert at End")
        print("3. Insert at Position")
        print("4. Delete at Beginning")
        print("5. Delete at End")
        print("6. Delete at Position")
        print("7. Update Node at Position")
        print("8. Print Linked List")
        print("9. Size of Linked List")
        print("10. Reverse Linked List")
        print("0. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            data = int(input("Enter data to insert at beginning: "))
            ll.insertAtBegin(data)
        elif choice == '2':
            data = int(input("Enter data to insert at end: "))
            ll.insertAtEnd(data)
        elif choice == '3':
            data = int(input("Enter data to insert: "))
            position = int(input("Enter position to insert at: "))
            ll.insertAtPosition(data, position)
        elif choice == '4':
            ll.deleteAtBegin()
        elif choice == '5':
            ll.deleteAtEnd()
        elif choice == '6':
            position = int(input("Enter position to delete: "))
            ll.deleteAtPosition(position)
        elif choice == '7':
            index = int(input("Enter position to update: "))
            val = int(input("Enter new value: "))
            ll.updateNode(val, index)
        elif choice == '8':
            ll.printLL()
        elif choice == '9':
            print(f"Size of Linked List: {ll.size()}")
        elif choice == '10':
            ll.reverseList()
        elif choice == '0':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter 0-9.")
```
